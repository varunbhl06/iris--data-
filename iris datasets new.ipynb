# matplotlib

```python
plt.plot(iris["SepalWidthCm"],iris["PetalLengthCm"],marker='*')
plt.show()
```

```python
plt.plot(iris["SepalWidthCm"],iris["PetalLengthCm"],marker='*',linestyle='none')
plt.show()
```

```python
plt.figure()
fig,ax=plt.subplots(1,2,figsize=(17, 9))
iris.plot(x="SepalLengthCm",y="SepalWidthCm",kind="scatter",ax=ax[0],sharex=False,sharey=False,label="sepal",color='r')
iris.plot(x="PetalLengthCm",y="PetalWidthCm",kind="scatter",ax=ax[1],sharex=False,sharey=False,label="petal",color='b')
ax[0].set(title='Sepal comparasion ', ylabel='sepal-width')
ax[1].set(title='Petal Comparasion',  ylabel='petal-width')
ax[0].legend()
ax[1].legend()
```

```python
iris.plot(kind='box', subplots=True, layout=(3,3), sharex=False, sharey=False)
plt.show()
```

```python
iris.PetalLengthCm.plot(kind='hist',figsize=(11,11))
plt.show()
```

```python
iris.PetalLengthCm.plot(kind='hist',figsize=(11,11),label="PetalLengthCm")
iris.SepalWidthCm.plot(kind='hist',figsize=(11,11),label="SepalLengthCm")
plt.legend()
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
iris.SepalWidthCm.plot(kind='line',color = "r",linewidth=2,grid=True,label="SepalWidthCm")
iris.PetalLengthCm.plot(kind='line',color = "b",linewidth=2,grid=True,label="PetalLengthCm")
plt.legend()
plt.xlabel("X")
plt.ylabel("y")
plt.show()

```

```python
plt.figure(figsize=(5,5))
plt.subplot(2,2,1)
iris.SepalLengthCm.plot(kind = 'box',figsize = (10,10))
plt.title('SepalLengthCm')
plt.show()
plt.subplot(2,2,2)
iris.SepalWidthCm.plot(kind = 'box',figsize = (10,10))
plt.title('SepalWidthCm')
plt.show()
plt.subplot(2,2,3)
iris.PetalLengthCm.plot(kind = 'box',figsize = (10,10))
plt.title('PetalLengthCm')
plt.show()
plt.subplot(2,2,4)
iris.PetalWidthCm.plot(kind = 'box',figsize = (10,10))
plt.title('PetalWidthCm')
plt.show()
```
